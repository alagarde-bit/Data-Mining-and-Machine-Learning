```
import numpy as np
import matplotlib.pyplot as plt
```

```
poptable = np.loadtxt('lab02_populations.csv.txt', dtype = int, delimiter=',', skiprows= 1)
print(poptable)
```

```
print('Rows:', poptable.shape[0])
print('Columns:', poptable.shape[1])
```

```
hares = poptable[:, 1]
print('Minimum number of Hares:', hares.min())
print('Maximum number of Hares:', hares.max())
print('Average number of Hares:', hares.mean())
```

```
lynxes = poptable[:, 2]
print('Minimum number of Lynxes:', lynxes.min())
print('Maximum number of Lynxes:', lynxes.max())
print('Average number of Lynxes:', lynxes.mean())
```

```
carrots = poptable[:, 3]
print('Minimum number of Carrots:', carrots.min())
print('Maximum number of Carrots:', carrots.max())
print('Average number of Carrots:', carrots.mean())
```

```
above_Hares = poptable[hares > 40000, 0]
print("Years when number of Hares surpassed 40000: ", above_Hares)
```

```
years = poptable[:, 0]
plt.plot(years, hares, label = 'Hares')
plt.plot(years, lynxes, label = 'Lynxes')
plt.plot(years, carrots, label = 'Carrots')
plt.legend(('Hares', 'Lynxes', 'Carrots'))
plt.ylabel('Population')
plt.xlabel('Year')
plt.title('Population Sizes Over Time')
plt.xticks(np.arange(years.min(), years.max() + 1, 5))
plt.show()
```

```
plt.hist(carrots, bins = 8)
plt.title('Distribution of The Number of Carrots')
plt.ylabel('Number of Years')
plt.xlabel('Year Population')
plt.show()
```

```
plt.scatter(hares, lynxes, marker = 'o', color = 'orange', alpha=0.4)
plt.xlabel('Hares Population')
plt.ylabel('Lynx Population')
plt.title('Relationship Between Hares and Lynx Populations')
plt.show()
```

There does not appear to be a positive or negative correlation between the hares and lynx populations.
